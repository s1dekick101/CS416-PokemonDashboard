```python
import pandas as pd
```

```python
pdx = pd.read_csv('PokeDex.csv')
pk = pd.read_csv('Pokemon.csv')
```

```python
pdx
```

```python
#Changes the # field in the pdx series from a float to int for later purposes pertaining to the hashmap creation
pdx = pdx.astype({"#": "int"})

#Handles the * that was assigned to some of the intial values within the CatchRate column 
for i in range(len(pdx['CatchRate'])):
    if '*' in pdx.at[i, 'CatchRate']:
        pdx.at[i, 'CatchRate'] = pdx.at[i, 'CatchRate'].replace('*', '')

pdx
```

```python
#Create Hashmap: # --> Key , CatchRate --> Value from pdx data 
#Assisgns the pokemon ID to its Catch Rate Value
pdxHashMap = {}

for i in range(len(pdx['#'])):
    pdxHashMap[pdx.at[i,'#']] = pdx.at[i,'CatchRate']
```

```python
pk
```

```python
#Creates new column inside of pk data set called "CatchRate" and intilizes all of its values to 0
pk['CatchRate'] = 0

#Appends the CatchRate Value based upon the pokemon ID to the pk(Pokemon.csv) data set
for i in range(len(pk)):
    for j in range(len(pdxHashMap)):
        if pk.at[i,'#'] == j:
              pk.at[i,'CatchRate'] = pdxHashMap[j]
```

```python
pk
```

```python
#Transfers all the data from the pk dataset to the Pokemon_All.csv 
pk.to_csv('Pokemon_All.csv', mode='a', columns=pk.columns, header=True)
```
